```python
#merged_player_position
#positions = [merged_player_position['Position_player'].strip()]
def clean_positions(position_string):
    
    cleaned_string = position_string.replace(', ', ',')

    # Replace '/' with ','
    cleaned_string = cleaned_string.replace('/', ',')

    # Replace ', and' with ','
    cleaned_string = cleaned_string.replace(', and', ',')
    
    cleaned_string = cleaned_string.replace('and', '')
    cleaned_string = cleaned_string.replace('  ', ',')
    cleaned_string = cleaned_string.replace(', ',',')
    
    #cleaned_string = position_string.replace(', ', ',')
    

    return cleaned_string
```

```python
from sklearn.linear_model import LinearRegression
#import statsmodels.api as sm
```

```python
reg_df = new_table[['FG', 'FT']].dropna()
```

```python
X = reg_df['FG'].values.reshape(-1,1)
Y = reg_df['FT'].values.reshape(-1,1)
reg = LinearRegression(fit_intercept=False).fit(X,Y)
y_pred = reg.predict(X)
plt.figure(figsize=(12, 6))
plt.scatter(X, Y)
plt.plot(X, y_pred, color='red')
plt.xlabel("Field Goal")
plt.ylabel("FT)")
plt.title('Field Goal vs FT - simple linear regression', fontsize=16);
```

```python
print(reg.coef_)
```

```python
reg_df_PTS = new_table[['Points_per_game', 'Minute_play_per_game']].dropna()
```

```python
X = reg_df_PTS ['Points_per_game'].values.reshape(-1,1)
Y = reg_df_PTS ['Minute_play_per_game'].values.reshape(-1,1)
reg_PTS = LinearRegression(fit_intercept=False).fit(X,Y)
y_pred = reg_PTS.predict(X)
plt.figure(figsize=(12, 6))
plt.scatter(X, Y)
plt.plot(X, y_pred, color='red')
plt.xlabel("pts")
plt.ylabel("Minute_per_game")
plt.title('PTS VS Minute per game - simple linear regression', fontsize=16);
```
